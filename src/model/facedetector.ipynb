```python
#libraries need to be imported
import cv2
from mtcnn.mtcnn import MTCNN
import os
```

```python
def getFaces(faceArray, result_list):
    #objective :detect and return faces present in 
    #function takes two parameter, first parameter is the numpyarray corresponds to image
    #second parameter , each index holds co-ordinates of faces(x,y,height,width) corresponds to 
    #eacg face present in image
    #Note : return list of faces
    
    # load the image
    # data = cv2.imread(filename)
    data = faceArray
    faces=[]
    for i in range(len(result_list)):
        # get coordinates
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        faces.append(data[y1:y2, x1:x2])
    return faces
```

```python
def savefaces(faces):
    #getting current working directory
    dir1 = os.getcwd()
    #saving each face in directory
    for i in range(len(faces)):
        temp="faces/"+"sample"+str(i)+".jpg"
        dirPath2 = os.path.join(dir1,temp)
        cv2.imwrite(dirPath2,faces[i])
```

```python
image_location = "/home/jeevesh/Desktop/SE/project/haridatt.jpg"
#provide file path
file_path=image_location
# load image from file
pixels = cv2.imread(file_path)
# create the detector, using default weights
detector = MTCNN()
# detect faces in the image
faces = detector.detect_faces(pixels)
# display faces on the original image
faces = getFaces(pixels, faces)
# faces is a list saving faces entries of faces present in image
# entry consist of  co-ordinate (x,y) along with height  and width

# savefaces is optional, it saves the faces in your local 
#Note: if you want to use this function
#there must be a folder with names"faces" in current working directory
savefaces(faces)
```
